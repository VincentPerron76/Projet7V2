```python
import pandas as pd
import numpy as np
from evidently.report import Report
from evidently.metric_preset import DataDriftPreset

# Charger les données
try:
    train_df = pd.read_csv("df_data_train_non_scalées.csv")
    test_df = pd.read_csv("df_data_test_non_scalées.csv")
except FileNotFoundError:
    print("Les fichiers CSV n'ont pas été trouvés. Veuillez vérifier les chemins.")
    exit()

# Supprimer la colonne TARGET du test set car elle n'existe pas et cause des erreurs.
if 'TARGET' in train_df.columns and 'TARGET' in test_df.columns:
    train_df = train_df.drop('TARGET', axis=1)
if 'TARGET' in train_df.columns and 'TARGET' not in test_df.columns:
    train_df = train_df.drop('TARGET', axis=1)
```

```python
data_drift_report = Report(metrics=[
    DataDriftPreset(),
])

data_drift_report.run(reference_data=train_df, current_data=test_df)

# Sauvegarder le rapport au format HTML
data_drift_report.save_html("data_drift_report.html")

# Afficher le rapport dans un notebook Jupyter (optionnel)
# data_drift_report.show()
```
